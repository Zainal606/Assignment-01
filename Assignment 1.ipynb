# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Count Plot using Sepal Width

```python
plt.figure(figsize=(8,5))
sns.countplot(x=data['SW'])
plt.xticks(rotation = 45)
plt.xlabel('Sepal Width')
plt.show()
```

# Scatter Plot using Petal Length vs Petal Width

```python
sns.scatterplot(x = 'PL', y = 'PW', data = data, hue = 'Classification')
plt.title('Petal Length vs Petal Width')
plt.show()
```

# Petal length vs Classification using barplot

```python
plt.bar(data['Classification'],data['PL'])
plt.xlabel('Classification')
plt.ylabel('Petal Length')
plt.title('Petal Length for diff classifiaction')
plt.show()
```
